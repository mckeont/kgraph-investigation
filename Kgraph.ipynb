```python
import json
from collections import defaultdict

# -----------------------------------------
# CONFIGURATION
# -----------------------------------------

MAX_HITS_PER_CHEMICAL = 20      # stop after this many hits per chemical
MAX_TOTAL_HITS = 200            # global stop condition
PROGRESS_INTERVAL = 100000      # print every N lines

# -----------------------------------------
# TRACKING STRUCTURES
# -----------------------------------------

results = []
count = 0

hits_per_chemical = defaultdict(int)
hits_per_target = defaultdict(int)

chemicals_remaining = set(chemical_ids)   # IDs of your 6 chemicals
targets_remaining = set(disease_ids | process_ids)

# -----------------------------------------
# STREAMING PARSER WITH EARLY STOPPING
# -----------------------------------------

with open("PubMedList.json", "r", encoding="utf-8") as f:
    for line in f:
        count += 1

        # Progress print
        if count % PROGRESS_INTERVAL == 0:
            print(f"Processed {count:,} lines | "
                  f"Total hits: {len(results)} | "
                  f"Remaining chemicals: {len(chemicals_remaining)}")

        # Stop early if enough evidence collected
        if len(results) >= MAX_TOTAL_HITS:
            print("\nStopping early: reached MAX_TOTAL_HITS")
            break

        line = line.strip()
        if not line:
            continue

        # Parse JSON safely
        try:
            obj = json.loads(line)
        except:
            continue

        # Skip non-dict entries
        if not isinstance(obj, dict):
            continue

        # Extract fields
        node1 = obj.get("Node1")
        node2 = obj.get("Node2")
        relation = obj.get("Relation")
        prob = obj.get("Probability")
        pmids = obj.get("PMIDs", [])

        if node1 is None or node2 is None:
            continue

        # Determine if this relationship involves one of your chemicals
        chem_side = None
        target_side = None

        if node1 in chemical_ids:
            chem_side = node1
            target_side = node2
        elif node2 in chemical_ids:
            chem_side = node2
            target_side = node1
        else:
            continue

        # Check if the target is relevant
        if target_side not in disease_ids and target_side not in process_ids:
            continue

        # Early stopping: skip if this chemical already has enough hits
        if hits_per_chemical[chem_side] >= MAX_HITS_PER_CHEMICAL:
            continue

        # Map IDs → names
        chem_name = id_to_name.get(chem_side, "UNKNOWN")
        target_name = id_to_name.get(target_side, "UNKNOWN")

        # Store result
        results.append({
            "chemical_id": chem_side,
            "chemical_name": chem_name,
            "target_id": target_side,
            "target_name": target_name,
            "relation": relation,
            "probability": prob,
            "pmids": pmids
        })

        # Update counters
        hits_per_chemical[chem_side] += 1
        hits_per_target[target_side] += 1

        # Remove from "remaining" sets
        if chem_side in chemicals_remaining:
            chemicals_remaining.remove(chem_side)

        if target_side in targets_remaining:
            targets_remaining.remove(target_side)

        # Stop early if all chemicals have at least one hit
        if len(chemicals_remaining) == 0:
            print("\nStopping early: all chemicals have hits")
            break

print("\nDone!")
print(f"Total relevant relationships found: {len(results)}")
print("Hits per chemical:", dict(hits_per_chemical))


```

```python
import json
from collections import defaultdict

# -----------------------------------------
# CONFIG
# -----------------------------------------

MAX_DOCS = 2_000_000          # safety cap on documents to scan
PROGRESS_INTERVAL = 10_000    # print every N documents
MAX_HITS = 500                # stop after this many chem–target hits

# -----------------------------------------
# STORAGE
# -----------------------------------------

results = []  # each: {chemical_id, chemical_name, target_id, target_name, pmid, prob}
hits_per_chemical = defaultdict(int)
hits_per_pair = defaultdict(int)

doc_count = 0

# -----------------------------------------
# STREAM THE BIG JSON ARRAY
# -----------------------------------------

with open("PubMedList.json", "r", encoding="utf-8") as f:
    # We can't json.load() the whole thing if it's huge, so we do a simple streaming parse:
    # read the file, strip the leading '[' and trailing ']', then split on '},' boundaries.
    # But safer: use a small state machine.

    buffer = ""
    inside_array = False
    brace_depth = 0

    while True:
        chunk = f.read(8192)
        if not chunk:
            break

        for ch in chunk:
            if not inside_array:
                if ch == '[':
                    inside_array = True
                continue

            # End of array
            if ch == ']':
                # flush last object if any
                if buffer.strip():
                    try:
                        obj = json.loads(buffer)
                    except:
                        obj = None
                    if isinstance(obj, dict):
                        # process last object
                        doc_count += 1
                        if doc_count % PROGRESS_INTERVAL == 0:
                            print(f"Processed {doc_count:,} documents | hits: {len(results)}")

                        # parse one document
                        doc_id = obj.get("id")
                        entries = obj.get("list", [])

                        # collect entities in this doc
                        chem_entities = []
                        target_entities = []

                        for entry in entries:
                            if not isinstance(entry, list) or len(entry) < 3:
                                continue
                            ent_id = entry[0]
                            pmid_str = entry[1]
                            prob = entry[2]

                            if ent_id in chemical_ids:
                                chem_entities.append((ent_id, pmid_str, prob))
                            if ent_id in target_ids:
                                target_entities.append((ent_id, pmid_str, prob))

                        # now pair chemicals and targets within this doc
                        for chem_id, pmid_c, prob_c in chem_entities:
                            for targ_id, pmid_t, prob_t in target_entities:
                                # they should share the same pmid; if not, we can still treat doc-level co-occurrence
                                pmid = pmid_c.split('.')[0]  # take the PubMed ID prefix
                                pair_key = (chem_id, targ_id, pmid)

                                if hits_per_pair[pair_key] > 0:
                                    continue  # avoid duplicates

                                chem_name = id_to_name.get(chem_id, "UNKNOWN")
                                targ_name = id_to_name.get(targ_id, "UNKNOWN")

                                results.append({
                                    "chemical_id": chem_id,
                                    "chemical_name": chem_name,
                                    "target_id": targ_id,
                                    "target_name": targ_name,
                                    "pmid": pmid,
                                    "probability_chem": prob_c,
                                    "probability_target": prob_t,
                                    "doc_id": doc_id,
                                })

                                hits_per_pair[pair_key] += 1
                                hits_per_chemical[chem_id] += 1

                                if len(results) >= MAX_HITS:
                                    print("\nReached MAX_HITS, stopping early.")
                                    inside_array = False
                                    break
                            if not inside_array:
                                break

                        if not inside_array or doc_count >= MAX_DOCS:
                            inside_array = False
                            break
                break  # done with file

            # Object boundary tracking
            if ch == '{':
                brace_depth += 1
            if brace_depth > 0:
                buffer += ch
            if ch == '}':
                brace_depth -= 1
                if brace_depth == 0:
                    # end of one object
                    obj_str = buffer
                    buffer = ""
                    try:
                        obj = json.loads(obj_str)
                    except:
                        obj = None

                    if not isinstance(obj, dict):
                        continue

                    doc_count += 1
                    if doc_count % PROGRESS_INTERVAL == 0:
                        print(f"Processed {doc_count:,} documents | hits: {len(results)}")

                    # parse one document
                    doc_id = obj.get("id")
                    entries = obj.get("list", [])

                    chem_entities = []
                    target_entities = []

                    for entry in entries:
                        if not isinstance(entry, list) or len(entry) < 3:
                            continue
                        ent_id = entry[0]
                        pmid_str = entry[1]
                        prob = entry[2]

                        if ent_id in chemical_ids:
                            chem_entities.append((ent_id, pmid_str, prob))
                        if ent_id in target_ids:
                            target_entities.append((ent_id, pmid_str, prob))

                    for chem_id, pmid_c, prob_c in chem_entities:
                        for targ_id, pmid_t, prob_t in target_entities:
                            pmid = pmid_c.split('.')[0]
                            pair_key = (chem_id, targ_id, pmid)

                            if hits_per_pair[pair_key] > 0:
                                continue

                            chem_name = id_to_name.get(chem_id, "UNKNOWN")
                            targ_name = id_to_name.get(targ_id, "UNKNOWN")

                            results.append({
                                "chemical_id": chem_id,
                                "chemical_name": chem_name,
                                "target_id": targ_id,
                                "target_name": targ_name,
                                "pmid": pmid,
                                "probability_chem": prob_c,
                                "probability_target": prob_t,
                                "doc_id": doc_id,
                            })

                            hits_per_pair[pair_key] += 1
                            hits_per_chemical[chem_id] += 1

                            if len(results) >= MAX_HITS:
                                print("\nReached MAX_HITS, stopping early.")
                                inside_array = False
                                break
                        if not inside_array:
                            break

                    if not inside_array or doc_count >= MAX_DOCS:
                        inside_array = False
                        break

print("\nDone.")
print(f"Documents processed: {doc_count}")
print(f"Total chem–target hits: {len(results)}")
print("Hits per chemical:", {id_to_name.get(k, k): v for k, v in hits_per_chemical.items()})
```

```python
import json

def read_json_object_at_offset(f, offset):
    # Move to the offset
    f.seek(offset)

    # Move backward until we hit '{'
    while True:
        f.seek(offset)
        ch = f.read(1)
        if ch == '{':
            break
        offset -= 1
        if offset < 0:
            return None

    # Now read forward until matching braces close
    brace_depth = 1
    obj_text = '{'
    while brace_depth > 0:
        ch = f.read(1)
        if not ch:
            return None
        obj_text += ch
        if ch == '{':
            brace_depth += 1
        elif ch == '}':
            brace_depth -= 1

    try:
        return json.loads(obj_text)
    except:
        return None
```
